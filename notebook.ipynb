# Reachability Ratio of Riga City Areas Using Public Transport

### Introduction
The Riga public transport network has not been fundamentally changed for over 30 years [1]. Urbanization and population trends, infrastructure, and habits, however, have changed [2]. Furthermore, improvements in the intercity and railway network, including the emergence of the Rail Baltica project, have increased pressure on the capital of Latvia to reconsider its current public transport network system.

### Reachability Ratio
In this project, the accessibility (reachability) of city areas from populated city areas is defined as
$$ \text{Reachability} = \frac{\text{city area reachable in one hour}}{\text{total city area}}$$

given that at least one public transport option is taken. By combining several open data sets of Latvia's Central Bureau of Statistics and Rigas Satiksme, a map of the reachability ratio for each $100m \times 100m$ city grid square was created.


![Reachability Ratio Map](https://raw.githubusercontent.com/oakw/riga-city-transport-reachability/master/reachability_ratio_map_en.png "Reachability ratio map")
![Reachability Ratio by Distance from City Center](https://raw.githubusercontent.com/oakw/riga-city-transport-reachability/master/reachability_ratio_en.png "Reachability Ratio by Distance from City Center")

Visualizations identify differences in reachability across different city areas. What's interesting is that even with relatively small distances from the city center, the difference in reachability can be substantial. For instance, 2km from the city center, the reachability ratio may range from 0.34 to 0.61. These findings, with additional analysis, may be used to develop recommendations for improvements in the public transport network.

### Context

In 2023, Riga City Council started work on the project "**Analysis of the Current Situation and Scenario Development for the Action Plan of the Public Transport Route Network Reform in Riga City**". At the beginning of 2025, the municipality procurement ended successfully with the sum agreed by contract of nearly 365 thousand euros [3].

Among other tasks, the contractor has to "Conduct an analysis of travel time (including cartographic analysis with isochrones) in relation to areas and services accessible by private transport from different parts of the city(..)" [3]

This project does not aim to fulfill this task or propose significant conclusions or suggestions. Rather, this is a **showcase of how one can use open data to analyze real-world problems and later visualize the results**.

Some potential **limitations and improvements**:
- The simulation of reachability does not consider natural or anthropogenic objects (e.g. river Daugava as a natural obstacle).
- Only a specific time is considered (Thursday, 8:00 on October 2024).
- Only Rigas Satiksme's public transport network is considered (inter-city busses or trains are not considered).
- Only official scheduled stop times were analyzed, not actual departure based on GPS data.
- Human habits are not much considered (the model assumes one would walk 4.8 km to the nearest bus stop)
- Improve: Look for population, not city area as a determinant of reachability.

## Preparing Transport Data

The basis of all concerned movement is Rigas Satiskme routes. We look at the route network as a network of stops. Let's load them.

```python
class Stop:
    """Class representing a public transport stop."""
    def __init__(self, stop_id, lat, lon, name = None):
        self.stop_id = stop_id
        self.lat = float(lat)
        self.lon = float(lon)
        self.name = name

    def __str__(self):
        return f"Stop {self.stop_id} ({self.name}): ({self.lat}, {self.lon})"

    def __repr__(self):
        return f"Stop('{self.stop_id}', {self.lat}, {self.lon})"

@dataclass
class StopWithDistance:
    """Class representing a Stop with its distance from a reference point."""
    stop: Stop
    distance: float  # distance in meters

    def __str__(self):
        return f"{self.stop}, Distance: {self.distance:.2f}m"

def load_stops(file_path):
    """Load stops from the stops.txt file."""
    stops = {}

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, fieldnames=['stop_id', 'stop_code',
              'stop_name', 'stop_desc', 'stop_lat', 'stop_lon', 'stop_url',
              'location_type', 'parent_station'])
            reader.__next__()  # Skip the header
            for row in reader:
                stop_id = row['stop_id']
                lat = row['stop_lat']
                lon = row['stop_lon']
                name = row['stop_name']

                if stop_id and lat and lon:
                    stops[stop_id] = Stop(stop_id, lat, lon, name)
    except Exception as e:
        print(f"Error loading stops: {e}")

    return stops

stops_dict = load_stops(STOPS_FILE)
print(f"Loaded {len(stops_dict)} stops")
```

The `stop_times.txt` file lists every departure from each of the stops. It is linked via `trips.txt` that lists all possible routes and `calendar.txt` that describes at which day which transport runs. All of this together makes the transport network.

```python
class Trip:
    """Class representing a trip from trips.txt."""
    def __init__(self, route_id, service_id, trip_id, trip_headsign=None, direction_id=None,
                 block_id=None, shape_id=None, wheelchair_accessible=None):
        self.route_id = route_id
        self.service_id = service_id
        self.trip_id = trip_id
        self.trip_headsign = trip_headsign
        self.direction_id = direction_id
        self.block_id = block_id
        self.shape_id = shape_id
        self.wheelchair_accessible = wheelchair_accessible

    def __str__(self):
        return f"Trip {self.trip_id}: {self.trip_headsign} (Service: {self.service_id})"

class Calendar:
    """Class representing a calendar entry from calendar.txt."""
    def __init__(self, service_id, monday, tuesday, wednesday, thursday, friday, saturday, sunday,
                 start_date, end_date):
        self.service_id = service_id
        self.days = [
            bool(int(monday)),
            bool(int(tuesday)),
            bool(int(wednesday)),
            bool(int(thursday)),
            bool(int(friday)),
            bool(int(saturday)),
            bool(int(sunday))
        ]
        self.start_date = self._parse_date(start_date)
        self.end_date = self._parse_date(end_date)

    def _parse_date(self, date_str):
        """Parse a date string in YYYYMMDD format."""
        if len(date_str) != 8:
            raise ValueError(f"Invalid date format: {date_str}")
        year = int(date_str[:4])
        month = int(date_str[4:6])
        day = int(date_str[6:8])
        return date(year, month, day)

    def is_service_available(self, target_date):
        """Check if service is available on the specified date."""
        # Check if the date is within the service period
        if not (self.start_date <= target_date <= self.end_date):
            return False

        # Check if service is available on the day of the week
        # weekday() returns 0 for Monday, 6 for Sunday
        return self.days[target_date.weekday()]

    def __str__(self):
        days_str = "MTWTFSS"
        days_status = ''.join(['•' if day else '·' for day in self.days])
        return f"Service {self.service_id}: [{days_status}] from {self.start_date} to {self.end_date}"

class TimeTableEntry:
    """Class representing a timetable entry."""
    def __init__(self, trip_id, arrival_time, departure_time, stop_id, stop_sequence):
        self.trip_id = trip_id

        # Parse time strings into timedelta objects for easier calculations
        self.arrival_time = self._parse_time(arrival_time)
        self.departure_time = self._parse_time(departure_time)

        self.stop_id = stop_id
        self.stop_sequence = int(stop_sequence)

    def _parse_time(self, time_str):
        """Parse a time string into a timedelta object."""
        # Handle times past midnight (e.g., 25:30:00)
        hours, minutes, seconds = map(int, time_str.split(':'))
        return timedelta(hours=hours, minutes=minutes, seconds=seconds)

    def __str__(self):
        return f"Trip {self.trip_id} at {self.departure_time} from stop {self.stop_id}"

class TransportNetwork:
    """Class representing the public transport network."""
    def __init__(self, stops_dict=None):
        self.stops = stops_dict or {}
        self.timetable = []
        self.trip_stops = defaultdict(list)  # trip_id -> ordered list of timetable entries
        self.stop_departures = defaultdict(list)  # stop_id -> list of timetable entries
        self.trips = {}  # trip_id -> Trip object
        self.calendar = {}  # service_id -> Calendar object
        self.active_services = set()  # Set of active service IDs for the selected date
        self.active_trip_ids = set()  # Set of active trip IDs for the selected date

        # Pre-compute nearby stops for each stop (within walking distance)
        self.nearby_stops = defaultdict(list)  # stop_id -> list of (stop_id, distance, walking_time)
```

Let's load the trips and their bindings to the calendar day. Move everything to the transport network.

```python
def load_trips(file_path):
    """Load trips from the trips.txt file."""
    trips = {}

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, fieldnames=['route_id', 'service_id', 'trip_id', 'trip_headsign', 'direction_id', 'block_id', 'shape_id', 'wheelchair_accessible'])
            reader.__next__()  # Skip the header
            for row in reader:
                trip_id = row['trip_id']
                route_id = row['route_id']
                service_id = row['service_id']
                trip_headsign = row.get('trip_headsign', None)
                direction_id = row.get('direction_id', None)
                block_id = row.get('block_id', None)
                shape_id = row.get('shape_id', None)
                wheelchair_accessible = row.get('wheelchair_accessible', None)

                trips[trip_id] = Trip(
                    route_id=route_id,
                    service_id=service_id,
                    trip_id=trip_id,
                    trip_headsign=trip_headsign,
                    direction_id=direction_id,
                    block_id=block_id,
                    shape_id=shape_id,
                    wheelchair_accessible=wheelchair_accessible
                )
    except Exception as e:
        print(f"Error loading trips: {e}")

    return trips


def load_calendar(file_path):
    """Load calendar data from the calendar.txt file."""
    calendar = {}

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, fieldnames=['service_id', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday', 'start_date', 'end_date'])
            reader.__next__()  # Skip the header
            for row in reader:
                service_id = row['service_id']
                monday = row['monday']
                tuesday = row['tuesday']
                wednesday = row['wednesday']
                thursday = row['thursday']
                friday = row['friday']
                saturday = row['saturday']
                sunday = row['sunday']
                start_date = row['start_date']
                end_date = row['end_date']

                calendar[service_id] = Calendar(
                    service_id=service_id,
                    monday=monday,
                    tuesday=tuesday,
                    wednesday=wednesday,
                    thursday=thursday,
                    friday=friday,
                    saturday=saturday,
                    sunday=sunday,
                    start_date=start_date,
                    end_date=end_date
                )
    except Exception as e:
        print(f"Error loading calendar: {e}")

    return calendar

def load_timetable(self, file_path):
    """Load timetable from stop_times.txt file."""
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            reader = csv.DictReader(f, fieldnames=['trip_id', 'arrival_time', 'departure_time', 'stop_id', 'stop_sequence', 'pickup_type', 'drop_off_type'])
            reader.__next__()  # Skip the header
            for row in reader:
                trip_id = row['trip_id']

                # Skip if the trip is not active for the selected date
                if self.active_trip_ids and trip_id not in self.active_trip_ids:
                    continue

                arrival_time = row['arrival_time']
                departure_time = row['departure_time']
                stop_id = row['stop_id']
                stop_sequence = row['stop_sequence']

                entry = TimeTableEntry(trip_id, arrival_time, departure_time, stop_id, stop_sequence)
                self.timetable.append(entry)
                self.trip_stops[trip_id].append(entry)
                self.stop_departures[stop_id].append(entry)

        # Sort the entries in each trip by stop sequence
        for trip_id, entries in self.trip_stops.items():
            entries.sort(key=lambda x: x.stop_sequence)

        # Sort the departures for each stop by departure time
        for stop_id, entries in self.stop_departures.items():
            entries.sort(key=lambda x: x.departure_time)

        print(f"Loaded {len(self.timetable)} timetable entries")
        print(f"Found {len(self.trip_stops)} trips with timetable entries")

    except Exception as e:
        print(f"Error loading timetable: {e}")

def load_trips_and_calendar(self, trips_file, calendar_file, target_date=None):
    """
    Load trips and calendar data and determine which services are active.

    Args:
        trips_file: Path to trips.txt
        calendar_file: Path to calendar.txt
        target_date: Date for which to determine active services (default: today)
    """
    self.trips = load_trips(trips_file)
    self.calendar = load_calendar(calendar_file)

    # Use today's date if no target date is provided
    if target_date is None:
        target_date = date.today()
    elif isinstance(target_date, str):
        # Parse date string in format YYYY-MM-DD
        target_date = datetime.strptime(target_date, "%Y-%m-%d").date()

    # Determine which services are active on the target date
    self.active_services = set()
    for service_id, calendar_entry in self.calendar.items():
        if calendar_entry.is_service_available(target_date):
            self.active_services.add(service_id)

    # Determine which trips are active on the target date
    self.active_trip_ids = set()
    for trip_id, trip in self.trips.items():
        if trip.service_id in self.active_services:
            self.active_trip_ids.add(trip_id)

    print(f"Loaded {len(self.trips)} trips and {len(self.calendar)} calendar entries")
    print(f"Active services for {target_date}: {len(self.active_services)}")
    print(f"Active trips for {target_date}: {len(self.active_trip_ids)}")

TransportNetwork.load_trips_and_calendar = load_trips_and_calendar
TransportNetwork.load_timetable = load_timetable

# Create the transport network
network = TransportNetwork(stops_dict)
day_name = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"][DATE_OF_INTEREST.weekday()]
print(f"Processing schedule for {day_name}, {DATE_OF_INTEREST}")
network.load_trips_and_calendar(TRIPS_FILE, CALENDAR_FILE, DATE_OF_INTEREST)
network.load_timetable(STOP_TIMES_FILE)
```

## Modeling Transport Network Traversal

Before moving forward to interconnected movement, let's precompute nearby stops one can reach:

```python
def compute_nearby_stops(self, max_walking_distance_meters=500, walking_speed_meters_per_minute=80):
    """
    Precompute nearby stops for each stop that are within walking distance.

    Args:
        max_walking_distance_meters: Maximum walking distance in meters
        walking_speed_meters_per_minute: Walking speed in meters per minute
    """
    print(f"Computing nearby stops (max distance: {max_walking_distance_meters}m, speed: {walking_speed_meters_per_minute}m/min)...")

    stop_list = list(self.stops.values())
    for i, stop1 in enumerate(stop_list):
        for stop2 in stop_list[i+1:]:
            distance = haversine_distance(stop1.lat, stop1.lon, stop2.lat, stop2.lon)

            if distance <= max_walking_distance_meters:
                walking_time = timedelta(minutes=distance / walking_speed_meters_per_minute)

                # Add each stop as nearby to the other
                self.nearby_stops[stop1.stop_id].append((stop2.stop_id, distance, walking_time))
                self.nearby_stops[stop2.stop_id].append((stop1.stop_id, distance, walking_time))

    # Sort nearby stops by distance
    for stop_id in self.nearby_stops:
        self.nearby_stops[stop_id].sort(key=lambda x: x[1])

    total_connections = sum(len(stops) for stops in self.nearby_stops.values())
    print(f"Computed {total_connections} walking connections between stops")

def haversine_distance(lat1: float, lon1: float, lat2: float, lon2: float) -> float:
    """
    Calculate the great circle distance between two points on the earth.
    Returns distance in meters.
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c = 2 * math.asin(math.sqrt(a))
    r = 6371000  # Earth radius in meters
    return c * r

TransportNetwork.compute_nearby_stops = compute_nearby_stops
network.compute_nearby_stops(MAX_WALKING_DISTANCE, WALKING_SPEED)
```

We will now model the interconnections of public transport and walking as a graph where nodes would be unique stops of public transport. These will later descibe all our movement possibilities.

```python
@dataclass
class TreeNode:
  """Node in the reachability tree."""
  stop_id: str
  arrival_time: timedelta
  parent: 'TransportNetwork.TreeNode' = None
  trip_id: str = None  # trip used to reach this stop or "walking" if walked
  children: List['TransportNetwork.TreeNode'] = None

  def __post_init__(self):
      if self.children is None:
          self.children = []

  def format_time(self, time_delta):
      """Format a timedelta as HH:MM:SS."""
      total_seconds = int(time_delta.total_seconds())
      hours = total_seconds // 3600
      minutes = (total_seconds % 3600) // 60
      seconds = total_seconds % 60
      return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

  def __str__(self):
      arrival_time_str = self.format_time(self.arrival_time)
      if self.parent:
          parent_id = self.parent.stop_id
          transport_mode = "walking" if self.trip_id == "walking" else f"trip {self.trip_id}"
          return f"Stop {self.stop_id} at {arrival_time_str} from {parent_id} via {transport_mode}"
      return f"Stop {self.stop_id} at {arrival_time_str}"

TransportNetwork.TreeNode = TreeNode
```

Now, using the [Breadth-first search](https://en.wikipedia.org/wiki/Breadth-first_search), we will start exploring from the closest stop further by:
- travelling using the public transport, then walking from any of the stops
- travelling using the public transport, then possibly walking, then taking a different public transport.

This process is allowed as long as one abides by the time limit.

```python
def find_nearest_stop(self, lat, lon, max_distance_meters=1000):
    """Find the nearest stop to a given coordinate."""
    nearest_stop = None
    min_distance = float('inf')

    for stop_id, stop in self.stops.items():
        distance = haversine_distance(lat, lon, stop.lat, stop.lon)
        if distance < min_distance and distance <= max_distance_meters:
            min_distance = distance
            nearest_stop = stop

    return nearest_stop, min_distance if nearest_stop else None

def build_reachability_tree_from_coords(self, start_lat, start_lon, start_time, time_limit_minutes=60,
                                        walking_speed_meters_per_minute=80, max_walking_distance_meters=500):
    """
    Build a reachability tree from a specific geographic coordinate.

    Args:
        start_lat: Starting latitude
        start_lon: Starting longitude
        start_time: Starting time as a string in format HH:MM:SS
        time_limit_minutes: Maximum travel time in minutes
        walking_speed_meters_per_minute: Walking speed in meters per minute
        max_walking_distance_meters: Maximum walking distance in meters

    Returns:
        Root node of the reachability tree
    """
    # Find the nearest stop to the starting point
    nearest_stop, distance_to_stop = self.find_nearest_stop(start_lat, start_lon)

    if not nearest_stop:
        # print(f"Error: No stops found within {max_walking_distance_meters}m of coordinates ({start_lat}, {start_lon})")
        return None

    # Calculate time to walk to the first stop
    walk_time_to_stop = timedelta(minutes=distance_to_stop / walking_speed_meters_per_minute)

    # Parse start time
    start_time_delta = self._parse_time(start_time)

    # Adjust start time to account for walking to the first stop
    adjusted_start_time = start_time_delta + walk_time_to_stop

    # Compute nearby stops if not done already
    if not self.nearby_stops:
        self.compute_nearby_stops(max_walking_distance_meters, walking_speed_meters_per_minute)

    return self.build_reachability_tree(nearest_stop.stop_id, self._format_timedelta(adjusted_start_time),
                                        time_limit_minutes - walk_time_to_stop.total_seconds() / 60)

def build_reachability_tree(self, start_stop_id, start_time, time_limit_minutes=60,
                          max_walking_distance_meters=500, walking_speed_meters_per_minute=80):
    """
    Build a tree of all stops reachable within the time limit, including walking between nearby stops.

    Args:
        start_stop_id: ID of the starting stop
        start_time: Starting time as a string in format HH:MM:SS
        time_limit_minutes: Maximum travel time in minutes
        max_walking_distance_meters: Maximum walking distance in meters
        walking_speed_meters_per_minute: Walking speed in meters per minute

    Returns:
        Root node of the reachability tree
    """
    if start_stop_id not in self.stops:
        print(f"Error: Stop {start_stop_id} not found")
        return None

    # Compute nearby stops if not done already
    if not self.nearby_stops:
        self.compute_nearby_stops(max_walking_distance_meters, walking_speed_meters_per_minute)

    # Parse start time
    start_time_delta = self._parse_time(start_time)
    time_limit_delta = timedelta(minutes=time_limit_minutes)
    end_time_delta = start_time_delta + time_limit_delta

    # Create root node
    root = self.TreeNode(stop_id=start_stop_id, arrival_time=start_time_delta)

    # Stores stops we've already visited at a better time
    visited = {}  # stop_id -> earliest arrival time
    visited[start_stop_id] = start_time_delta

    # BFS queue for traversal
    queue = deque([root])

    while queue:
        node = queue.popleft()
        current_stop_id = node.stop_id
        current_time = node.arrival_time

        # Skip if we've reached our time limit
        if current_time > end_time_delta:
            continue

        # OPTION 1: Take public transport from this stop
        # Find all departures from this stop after current time
        departures = [entry for entry in self.stop_departures[current_stop_id]
                      if entry.departure_time >= current_time]

        for departure in departures:
            trip_id = departure.trip_id
            trip_stops = self.trip_stops[trip_id]

            # Find the current stop in the trip
            for i, entry in enumerate(trip_stops):
                if entry.stop_id == current_stop_id and entry.departure_time >= current_time:
                    # Look at all subsequent stops in this trip
                    for next_entry in trip_stops[i+1:]:
                        next_stop_id = next_entry.stop_id
                        arrival_time = next_entry.arrival_time

                        # Skip if we would exceed our time limit
                        if arrival_time > end_time_delta:
                            break

                        # Skip if we've already reached this stop earlier or at the same time
                        if next_stop_id in visited and visited[next_stop_id] <= arrival_time:
                            continue

                        # Create a new node and add it to the tree
                        child = self.TreeNode(
                            stop_id=next_stop_id,
                            arrival_time=arrival_time,
                            parent=node,
                            trip_id=trip_id
                        )
                        node.children.append(child)

                        # Mark this stop as visited with this arrival time
                        visited[next_stop_id] = arrival_time

                        # Add to the queue for further exploration
                        queue.append(child)

                    # We found the right entry in this trip, no need to check others
                    break

        # OPTION 2: Walk to nearby stops
        for nearby_stop_id, distance, walking_time in self.nearby_stops[current_stop_id]:
            nearby_arrival_time = current_time + walking_time

            # Skip if we would exceed our time limit
            if nearby_arrival_time > end_time_delta:
                continue

            # Skip if we've already reached this stop earlier or at the same time
            if nearby_stop_id in visited and visited[nearby_stop_id] <= nearby_arrival_time:
                continue

            # Create a new node for the walk and add it to the tree
            child = self.TreeNode(
                stop_id=nearby_stop_id,
                arrival_time=nearby_arrival_time,
                parent=node,
                trip_id="walking"  # Special identifier for walking
            )
            node.children.append(child)

            # Mark this stop as visited with this arrival time
            visited[nearby_stop_id] = nearby_arrival_time

            # Add to the queue for further exploration
            queue.append(child)

    return root

def _parse_time(self, time_str):
    """Parse a time string into a timedelta object."""
    hours, minutes, seconds = map(int, time_str.split(':'))
    return timedelta(hours=hours, minutes=minutes, seconds=seconds)

def _format_timedelta(self, td):
    """Format a timedelta as HH:MM:SS."""
    total_seconds = int(td.total_seconds())
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    seconds = total_seconds % 60
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

TransportNetwork.find_nearest_stop = find_nearest_stop
TransportNetwork.build_reachability_tree_from_coords = build_reachability_tree_from_coords
TransportNetwork.build_reachability_tree = build_reachability_tree
TransportNetwork._parse_time = _parse_time
TransportNetwork._format_timedelta = _format_timedelta
```

## Isochrone Area

Next, let's find out the the area one can get to within the timeframe, also called isochrone. The attainable area is defined as a region a person can achieve on foot, by transport, or any combination of both.

```python
def create_attainable_area(self, root_node, buffer_distance=200, walking_speed_meters_per_minute=80):
    """
    Create a polygon representing the area that can be reached within the time limit.

    Args:
        root_node: Root node of the reachability tree
        buffer_distance: Maximum buffer distance in meters for the walking radius around each stop
        walking_speed_meters_per_minute: Walking speed in meters per minute

    Returns:
        GeoDataFrame containing the polygon of the attainable area
    """
    # First, collect all reachable stops and their arrival times
    reachable_stops = {}  # stop_id -> arrival_time
    start_time = root_node.arrival_time

    def collect_stops(node):
        if node.stop_id in self.stops:
            # Only update if this is a better (earlier) arrival time
            if node.stop_id not in reachable_stops or node.arrival_time < reachable_stops[node.stop_id]:
                reachable_stops[node.stop_id] = node.arrival_time
        for child in node.children:
            collect_stops(child)

    collect_stops(root_node)

    # Create points for all reachable stops with remaining time buffers
    stop_points = []
    buffer_sizes = []

    for stop_id, arrival_time in reachable_stops.items():
        stop = self.stops[stop_id]

        # Calculate time remaining at this stop (in minutes)
        # Since we're using a time-based limit, the buffer size is adjusted by remaining time
        time_elapsed = arrival_time - start_time
        total_limit = root_node.format_time(root_node.arrival_time + timedelta(minutes=60))
        total_limit_delta = self._parse_time(total_limit) # Assuming 60 minutes is our limit
        time_remaining_seconds = (total_limit_delta - arrival_time).total_seconds()

        if time_remaining_seconds <= 0:
            # No time left for walking from this stop
            continue

        # Convert remaining time to maximum walking distance
        time_remaining_minutes = time_remaining_seconds / 60
        remaining_walk_distance = time_remaining_minutes * walking_speed_meters_per_minute

        # Use the smaller of the maximum buffer and what can be reached in remaining time
        actual_buffer = min(buffer_distance, remaining_walk_distance)

        stop_points.append(Point(stop.lon, stop.lat))
        buffer_sizes.append(actual_buffer)

    if not stop_points:
        print("No reachable stops found")
        return None

    # Create a GeoDataFrame with the stop points
    stops_gdf = gpd.GeoDataFrame({'buffer': buffer_sizes}, geometry=stop_points)

    # Set the CRS to WGS84 (standard for GPS coordinates)
    stops_gdf = stops_gdf.set_crs(epsg=4326)

    # Convert to a projected CRS for accurate distance calculations (meters)
    # Using UTM zone for Latvia (EPSG:3059)
    stops_gdf = stops_gdf.to_crs(epsg=3059)

    # Buffer each stop by its calculated buffer size
    buffered_geometries = [
        geom.buffer(buffer)
        for geom, buffer in zip(stops_gdf.geometry, stops_gdf['buffer'])
    ]

    # Merge all the buffered areas
    attainable_area = unary_union(buffered_geometries)

    # Create a GeoDataFrame with the merged area
    area_gdf = gpd.GeoDataFrame(geometry=[attainable_area], crs=stops_gdf.crs)

    # Convert back to WGS84 for general compatibility
    area_gdf = area_gdf.to_crs(epsg=4326)

    return area_gdf

def create_isochrone_map(self, root_node, grid_size=500, buffer_distance=200, walking_speed_meters_per_minute=80):
    """
    Create an isochrone map showing travel time to different areas.

    Args:
        root_node: Root node of the reachability tree
        grid_size: Size of the grid for interpolation
        buffer_distance: Maximum buffer distance in meters for the walking radius
        walking_speed_meters_per_minute: Walking speed

    Returns:
        Tuple of (gridx, gridy, travel_times) for plotting
    """
    # Collect all reachable stops and their travel times
    start_time = root_node.arrival_time
    total_time_limit = 60  # Assuming 60 minutes maximum
    reachable_stops = {}  # stop_id -> arrival_time

    def traverse_tree(node):
        if node.stop_id in self.stops:
            # Only update if this is a better (earlier) arrival time
            if node.stop_id not in reachable_stops or node.arrival_time < reachable_stops[node.stop_id]:
                reachable_stops[node.stop_id] = node.arrival_time

        for child in node.children:
            traverse_tree(child)

    traverse_tree(root_node)

    # Extract coordinates, travel times, and remaining times
    lats = []
    lons = []
    travel_minutes = []
    remaining_minutes = []

    for stop_id, arrival_time in reachable_stops.items():
        stop = self.stops[stop_id]

        # Calculate travel time and remaining time
        time_elapsed = arrival_time - start_time
        elapsed_minutes = time_elapsed.total_seconds() / 60
        remaining_time = total_time_limit - elapsed_minutes

        if remaining_time <= 0:
            # No time left for walking from this stop
            continue

        lats.append(stop.lat)
        lons.append(stop.lon)
        travel_minutes.append(elapsed_minutes)
        remaining_minutes.append(remaining_time)

    if not lats:
        print("No reachable stops found")
        return None

    # Create a grid for interpolation
    lat_min, lat_max = min(lats) - 0.01, max(lats) + 0.01
    lon_min, lon_max = min(lons) - 0.01, max(lons) + 0.01

    grid_x, grid_y = np.mgrid[lon_min:lon_max:complex(0, grid_size),
                            lat_min:lat_max:complex(0, grid_size)]

    # Interpolate initial travel times onto the grid
    points = np.column_stack((lons, lats))
    grid_times = griddata(points, travel_minutes, (grid_x, grid_y), method='linear')

    # For each grid point, check if walking from a stop would be faster
    for i in range(grid_size):
        for j in range(grid_size):
            grid_point_lon, grid_point_lat = grid_x[i, j], grid_y[i, j]
            best_time = np.inf if np.isnan(grid_times[i, j]) else grid_times[i, j]

            # Check all stops to see if walking from them would be faster
            for idx, stop_id in enumerate(reachable_stops):
                stop = self.stops[stop_id]
                distance = haversine_distance(
                    grid_point_lat, grid_point_lon, stop.lat, stop.lon
                )

                # Calculate walking time from this stop to the grid point
                walking_time = distance / walking_speed_meters_per_minute

                # Only consider points that can be reached within the remaining time
                if walking_time <= remaining_minutes[idx]:
                    # Total time = time to reach the stop + time to walk from stop to grid point
                    total_time = travel_minutes[idx] + walking_time

                    # Update if this is faster than current best time
                    if total_time < best_time:
                        best_time = total_time

            if best_time < np.inf:
                grid_times[i, j] = best_time

    return grid_x, grid_y, grid_times

TransportNetwork.create_attainable_area = create_attainable_area
TransportNetwork.create_isochrone_map = create_isochrone_map
```

## Comprehensive Analysis

In a similar manner, the total attainable area is found for each of the popoulated area:


```python
# Find midpoints for each of the populated gird square
points_of_interest = {}
riga_populated_grid_wgs84 = riga_populated_grid.to_crs(epsg=4326)
for idx, row in riga_populated_grid_wgs84.iterrows():
    # Get the centroid of the grid cell geometry
    centroid = row.geometry.centroid
    grid_id = row['grid_id']

    # Store as [longitude, latitude]
    points_of_interest[grid_id] = [centroid.y, centroid.x]

areas_of_interest = {}
def process_point(grid_id, point):
    start_lat, start_lon = point
    city_traversal = network.build_reachability_tree_from_coords(
        start_lat, start_lon, TIME_OF_INTEREST, TIME_LIMIT, WALKING_SPEED, MAX_WALKING_DISTANCE
    )
    if city_traversal:
        return grid_id, network.create_attainable_area(city_traversal, MAX_WALKING_DISTANCE, WALKING_SPEED)
    return None

areas_of_interest = {}

with ThreadPoolExecutor(max_workers=4) as executor:
    futures = {executor.submit(process_point, grid_id, point): grid_id for grid_id, point in points_of_interest.items()}

    for future in tqdm(as_completed(futures), total=len(futures), desc="Extracting attainable areas"):
        result = future.result()
        if result:
            grid_id, area = result
            areas_of_interest[grid_id] = area
```

Now, let's assume that the accessibility of the city is determined by how large the attainable area is. Then for each of the populated grid one can compute the accessibility ratio. Then it is displayed on a map.

```python
reachability_ratios = {}
riga_populated_grid_web_mercator = riga_populated_grid.to_crs(epsg=3857)
riga_boundaries_web_mercator = riga_boundaries.to_crs(epsg=3857)

total_riga_area = riga_boundaries_web_mercator.geometry.union_all().area

for grid_id, area_gdf in areas_of_interest.items():
    if area_gdf is not None:
        area_gdf_web_mercator = area_gdf.to_crs(epsg=3857)

        # Calculate the intersection with riga boundaries for accurate ratio
        intersection = area_gdf_web_mercator.geometry.intersection(riga_boundaries_web_mercator.geometry.union_all())
        reachability_ratios[grid_id] = (intersection.area / total_riga_area)[0]

riga_populated_grid['reachability_ratio'] = None
riga_populated_grid['reachability_ratio'] = riga_populated_grid['grid_id'].map(reachability_ratios)
riga_populated_grid.dropna(subset=['reachability_ratio'], inplace=True)

fig, ax = plt.subplots(figsize=(12, 10))

riga_boundaries_web_mercator.plot(ax=ax, color='none', edgecolor='black', linewidth=2, alpha=0.8)
riga_populated_grid_web_mercator.plot(ax=ax, column='reachability_ratio', cmap='viridis_r', edgecolor='gray', linewidth=0.1, alpha=0.7, legend=True)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, zoom=12, attribution_size=8)

ax.set_title('Reachability Ratio of Populated Riga City Areas', fontsize=16)
ax.annotate("Note: Relative measure of city area one can reach within an hour using at least one public transport option.\n"
            "Example of 8:00 am Thursday of December 2024. Source: Central Bureau of Statistics of Latvia, Rigas Satiksme.",
            xy=(0, -0.05), xycoords='axes fraction',
            ha="left", fontsize=8, style='italic', linespacing=2)
plt.axis('off')
plt.show()
```

It can be observed that, if starting the iterary from the city center, one can reach way more area in the city within one hour. While this makes sense given the geographical position of the city center, it is worth exploring whether equidistanly from the city center, the accessibility coefficient remains the same.
